```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Age'], y=df['Frequent Cold'])
plt.title("Distribution of Age vs Frequent Cold")
plt.xlabel("Age")
plt.ylabel("Frequent Cold")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Obesity'],  patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title("Outlier Detection in Obesity Feature")
plt.show()
```
